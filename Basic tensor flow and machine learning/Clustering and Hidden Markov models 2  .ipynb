```python
# Hidden Markov Models work with probability distributions
# Use predfined probability distributions associated with a finite set of states
# Transitions among the states are governed by transition probabilities
# Hidden (because we dont direcly observe) states have observations (probability of something happening during that state)
# dont need a lot of data to train the model, only need constanta values for probability of transition distribution
# Each state have a probability to transitioning to a different state
# The data can be standar deviation (probablities distribution) or simply probabilities percents
# The point of this model is predict future events based on past events
```

```python
import tensorflow_probability as tfp  # We are using a different module from tensorflow this time
import tensorflow as tf
```

```python
tfd = tfp.distributions  # making a shortcut for later on
initial_distribution = tfd.Categorical(probs=[0.2, 0.8])  # Refer to point 2 above
transition_distribution = tfd.Categorical(probs=[[0.7, 0.3],
                                                 [0.2, 0.8]])  # refer to points 3 and 4 above
observation_distribution = tfd.Normal(loc=[0., 15.], scale=[5., 10.])  # refer to point 5 above

# the loc argument represents the mean and the scale is the standard devitation
```

```python
model = tfd.HiddenMarkovModel(
    initial_distribution=initial_distribution,
    transition_distribution=transition_distribution,
    observation_distribution=observation_distribution,
    num_steps=7)
```

```python
mean = model.mean() #takes mean value from the model like a Tensor (partially defined computations)

# due to the way TensorFlow works on a lower level we need to evaluate part of the graph
# from within a session to see the value of this tensor

# in the new version of tensorflow we need to use tf.compat.v1.Session() rather than just tf.Session()
with tf.compat.v1.Session() as sess:  
  print(mean.numpy())
```
